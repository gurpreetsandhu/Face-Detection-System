### Training

```python
def read_img(path):
    return cv2.imread(pic,cv2.IMREAD_UNCHANGED)

def resize_img(img, width, height):
    dim = (width, height)
    return cv2.resize(img, dim, interpolation = cv2.INTER_AREA)

def rotate_img(img):
    return cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

def write_img(name, img):
    cv2.imwrite(name, img)
    
def create_dir(path):
    if not os.path.exists(path):
        os.makedirs(path)
```

```python
path = r"E:\Sem3\MC\Project\Dataset_used\input\train"
out = r"E:\Sem3\MC\Project\Dataset_used\output"

inp = path.split("\\")

result_folder = os.path.join(out, inp[-1])
print(result_folder)

create_dir(result_folder)
detector = MTCNN()

for cls in glob.glob(os.path.join(path, "*")):
    cls_name = cls.split("\\")[-1]
    print('\n--------------------------------------------------')
    print(cls_name)
    cls_folder = os.path.join(result_folder, cls_name)
#     print(cls_folder)
    create_dir(cls_folder)
    
    counter = 1
    img_count = 1
    for pic in glob.glob(os.path.join(cls, "*")):
#         print(pic)
#         img = cv2.imread(pic, cv2.IMREAD_UNCHANGED)
#         width, height = 1280, 960
#         resized_img = resize_img(img, width, height)
#         rotated_img = rotate_img(resized_img)

        img_name = pic.split("\\")[-1]
        print('\n',img_count,'::',img_name,end='');
        img_count += 1
        
        extension = img_name.split(".")[-1]

        img = cv2.imread(pic)
        result_list = detector.detect_faces(img)
        ax = plt.gca()
#         plt.imshow(img)
        if(len(result_list)==0):
            print(' | No |',len(result_list),end=' | ')
        else:
            print(' | Yes |',len(result_list),end=' | ')
        for result in result_list:
            x, y, width, height = result['box']
            rect = patches.Rectangle((x, y), width, height, fill=False, color='red')
#             ax.add_patch(rect)
#             plt.show()
            
            new_img = img[y:y+height , x:x+width, :]
#             plt.imshow(new_img)
#             plt.show()
            new_img_name = cls_name + "_" + str(counter) + "." + extension
            print(new_img_name,end='--')
            new_path = os.path.join(cls_folder, new_img_name)
            counter += 1
            write_img(new_path ,new_img)
        
#         plt.show()
```
